```python
import pandas as pd
import pickle
```

# 파일에서 데이터 각각 불러오기

```python
with open("./dataset/batting.p", "rb") as f:
    batting = pickle.load(f)
with open("./dataset/fa.p", 'rb') as f:
    fa = pickle.load(f)
with open("./dataset/fielding.p", 'rb') as f:
    fielding = pickle.load(f)
with open("./dataset/pitching.p", 'rb') as f:
    pitching = pickle.load(f)
```

# fa 데이터를 데이터프레임 형태로 가공

```python
li = []
```

```python
for i in fa:
    for j in fa[i]:
        li.append([i, *j])
```

```python
df = pd.DataFrame(li)
df.columns = ["name", "season", "age", "contract"]
```

# 데이터들을 리스트화 시킬 함수 생성

```python
def make_list(l):
    ll = []
    for i in l:
        for j in l[i]:
            ll.append([i, *j])
    return ll
```

```python
batting = make_list(batting)
fielding = make_list(fielding)
pitching = make_list(pitching)
```

# 생성한 3개의 리스트를 war이라는 하나의 리스트로 합친 후 데이터 프레임 형태로 가공

```python
war = batting+fielding+pitching
```

```python
war_df = pd.DataFrame(war)
```

```python
war_df.columns = ["name", "season", "WAR"]
```

# 이름들만 따로 빼서 저장
## set 자료형으로 저장해서 중복제거

```python
name = set(war_df['name'])
```

```python
data = []
for i in name:
    name_li = war_df.loc[war_df['name']==i]
    season = set(name_li['season'])
    for j in season:
        season_li = name_li.loc[name_li['season']==j]
        data.append([i, j, round(sum(map(float, season_li['WAR'])), 1)])
print(data)
```

```python
data_df = pd.DataFrame(data)
```

```python
fa = []
for i in name:
    name_li = df.loc[df['name']==i]
    season = set(name_li['season'])
    for j in season:
        season_li = name_li.loc[name_li['season']==j]
        t = 0
        temp = season_li['contract'].tolist()
        for k in temp:
            t = t or k
        fa.append([i, j, season_li['age'].iloc[0], t])
```

```python
for i in data:
    if i[1] == '2020':
        i[3] = round(i[3]*2.7, 1)
```

```python
data_df = pd.DataFrame(data)
data_df.columns = ["name", "season", "WAR"]
fa_df = pd.DataFrame(fa)
fa_df.columns = ["name", "season", "age", "contract"]
```

```python
df = []
for i in data:
    a = fa_df.loc[(fa_df["name"] == i[0]) & (fa_df["season"]==i[1])].values.tolist()
    if len(a) == 0:
        continue
    df.append([a[0][0], a[0][1], a[0][2], i[2], a[0][3]])
```

```python
dataset = []
for i in name:
    temp_name = sorted(df[df["name"]==i].values.tolist())
    comp_name = data_df_dict[i]
    for j in range(len(temp_name)):
        li = list(filter(lambda x: x < temp_name[j][1], comp_name.keys()))
        if len(li) == 0:
            continue
        s = 0
        for k in li:
            s += data_df_dict[i][k]
        dataset.append([i, temp_name[j][1]-2002, float(temp_name[j][2]), round(s, 1), data_df_dict[i][max(li)], temp_name[j][4]])
```
